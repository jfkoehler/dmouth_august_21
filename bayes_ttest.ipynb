### Bayes t-test with `bambi`

```python
import bambi as bmb
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/marketing.csv')
```

```python
df.head()
```

```python
import numpy as np
```

```python
g1 = np.random.normal(10, 3, 50)
g2 = np.random.normal(20, 3, 50)
df = pd.DataFrame({'SalesInThousands': np.hstack([g1, g2]), 'Promotion': ['p1']*50 + ['p2']*50})
df.head()
```

```python
model = bmb.Model("SalesInThousands ~ Promotion", df)
```

```python
results = model.fit()
```

```python
import arviz as az
```

```python
model.plot_priors()
```

```python
az.plot_trace(results)
```
