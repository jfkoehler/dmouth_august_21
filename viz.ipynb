```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
x = np.arange(-10*np.pi, 10*np.pi, 0.02)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.arctan(x))
plt.xlabel("x-coordinates")
plt.ylabel("y-coordinates")
plt.title("Some Trig Functions");
```

```python
np.random.seed(38472)
x = np.random.random(15)
y = np.random.random(15)

fig, ax = plt.subplots(1, 2)
ax[0].scatter(x, y, marker = 'o', s = 90)
ax[1].scatter(x, y, marker = '+', s = 90)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].scatter(x, y, marker = 'o', s = 90, c = np.sqrt(x**2 + y**2))
ax[1].scatter(x, y, marker = '+', s = 90, c = np.sqrt(x**2 + y**2))
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style = 'ticks', palette = 'pastel')
```

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'time')
```

```python
sns.catplot(data = tips, x = 'tip', y = 'day', 
            hue = 'time', col ='smoker', orient = 'h',kind = 'box')
```

```python
sns.catplot(data = tips, x = 'total_bill', y = 'day', 
            hue = 'time', col ='smoker', orient = 'h',kind = 'box')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.set(style = 'darkgrid')
sns.regplot(data = titanic, x = 'fare', y = 'age')
```
