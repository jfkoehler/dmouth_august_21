```python
np.random.seed(42)
for i in range(17):
    print(np.random.choice(coins))
```

```python
def run_experiment():
    result = ''
    for i in range(17):
        result += np.random.choice(coins)
    return result.count('H')
```

```python
experiment_1 = run_experiment()
```

```python
experiment_1.count('H')
```

```python
np.random.seed(42)
x = []
for exp in range(70):
    x.append(run_experiment())
```

**Math Quiz**

There are two bags. The first bag contains four mangos and two apples; the second bag contains four mangos and four apples. We also have a biased coin, which shows heads with probability 0.6 and tails with probability 0.4.

If the coin shows heads, we pick a fruit at random from bag 1; otherwise we pick a fruit at random from bag 2. Your friend flips the coin (you cannot see the result), picks a fruit at random from the corresponding bag, and presents you a mango.

What is the probability that the mango was picked from bag 2?

$${\displaystyle P(A\mid B)={\frac {P(B\mid A)P(A)}{P(B)}}}$$

$$g(x) = 3x^{-2} \text{on} ~  1 < x < 2 $$
