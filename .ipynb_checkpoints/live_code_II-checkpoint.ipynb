### Data EDA

```python
hungary = pd.read_csv('data/hungary_chickenpox.csv')
```

```python
hungary.info()
```

```python
hungary['BUDAPEST'].plot()
```

```python
hungary = hungary.set_index(pd.date_range(start = '01-01-2005', end = '12-31-2014', freq = 'W'))
```

```python
hungary['BUDAPEST'].plot()
```

```python
hungary.groupby(hungary.index.month).sum().plot(legend = False)
```

### Example: Moving Average

```python
hungary['BUDAPEST'].rolling(4)
```

```python
hungary['BUDAPEST'].rolling(4).mean()
```

```python
hungary['BARANYA'].rolling(13).mean().plot()
```

```python
hungary.sum(axis = 1).rolling(4).mean().plot()
```
