```python
def assert_normality(col):
    stat, p = shapiro(col)
    return p < 0.5
    
```

```python
df_norm = df.select_dtypes(['int', 'float']).apply(yeojohnson)
```

```python
def norm_maker(col):
    return yeojohnson(col)[0]
```

```python
df_norm = df.select_dtypes(['int', 'float']).apply(norm_maker)
```

```python
df_norm.info()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
df['AveragePrice'].hist(ax = ax[0])
df_norm['AveragePrice'].hist(ax = ax[1])
```

```python
X = df_norm.drop("AveragePrice", axis = 1)
```

```python
y = df_norm['AveragePrice']
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
rfe = RFE(LinearRegression(), n_features_to_select=5)
```

```python
X
```

```python
XT = rfe.fit_transform(X, y)
```

```python
rfe.get_support()
```

```python
lr = LinearRegression()
lr.fit(XT, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(XT, y)
```

```python
lr.fit(X_train, y_train)
```

```python
preds = lr.predict(X_test)
```

```python
print(f'''MAE: {mean_absolute_error(y_test, preds)}
MSE: {mean_squared_error(y_test, preds)}
RMSE:{mean_squared_error(y_test, preds, squared = False)}''')
```

```python
df['AveragePrice'].mean()
```

```python
lr.score(X_test, y_test)
```

### Doing it with a `pipeline`

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.pipeline import make_pipeline
```

```python
dft = df.select_dtypes(['int', 'float'])
```

```python
X = dft.drop('AveragePrice', axis = 1)
y = dft['AveragePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pipe = make_pipeline(PowerTransformer(), 
                     RFE(LinearRegression(), n_features_to_select=5),
                     LinearRegression())
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
[i for i,j in zip(dft.columns, pipe.named_steps['rfe'].get_support()) if j]
```

```python
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
selector = make_column_transformer((PowerTransformer(), 
                                    make_column_selector(dtype_include=np.number)),
                                  (OneHotEncoder(), 
                                  make_column_selector('type')))
```

```python
xt = selector.fit_transform(df)
```
