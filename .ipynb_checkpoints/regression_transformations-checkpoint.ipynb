### Regression: Extending the Models

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def f(x): return 50*x**2  + np.random.normal(loc = 0, scale = 10, size = 1000)
x = np.linspace(0, 3, 1000)
```

```python
plt.scatter(x, f(x))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x.reshape(-1, 1), f(x))
```

```python
preds = lr.predict(x.reshape(-1, 1))
plt.hist(f(x) - preds)
```

```python
plt.plot(x, preds, '--ro')
plt.scatter(x, f(x))
```

**MODEL**: $y = \beta_0 + \beta_1*x$ or maybe $y = \beta_0 + \beta_1*x + \beta_2*x^2$

```python
X = pd.DataFrame({'x': x, 'x^2': x**2})
```

```python
lr.fit(X, f(x))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
def f(x): return 50*x**2  + np.random.normal(loc = 0, scale = 10, size = 1000)
```

```python
quad_preds = lr.predict(X)
```

```python
plt.hist(f(x) - quad_preds);
```

```python
plt.plot(x, preds, '--ro')
plt.plot(x, quad_preds, '--go')
plt.scatter(x, f(x))
```
