```python
from sklearn.linear_model import LogisticRegression
```

```python
import statsmodels.api as sm
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(LogisticRegression(), n_features_to_select=8)
```

```python
XT = rfe.fit_transform(X_scaled, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(XT, y, test_size=.2)
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(X_train, y_train)
```

```python
lgr.score(X_test, y_test)
```

```python
lgr.score(X_train, y_train)
```

```python
from sklearn.metrics import plot_roc_curve, RocCurveDisplay
```

```python
RocCurveDisplay.from_estimator(lgr, X_test, y_test)
```

```python
plot_roc_curve(lgr, X_test, y_test)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(MaxAbsScaler(), 
                     RFE(LogisticRegression(), n_features_to_select=8),
                     LogisticRegression())
```

```python
pipe.fit(XT, y)
```

```python
pipe.score(XT, y)
```
